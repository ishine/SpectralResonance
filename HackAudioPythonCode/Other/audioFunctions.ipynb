```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io.wavfile import read, write
from IPython.display import Audio
import librosa
import soundfile
from numpy.fft import fft, ifft
```

```python
Fs, data = read('sw20.wav')

print('Sampling Frequency is: ', Fs)
```

```python
Audio(data, rate=Fs)
```

```python
plt.figure()
plt.plot(data)
plt.xlabel('Sample Index')
plt.ylabel('Amplitude')
plt.title('Waveform of Test Audio')
plt.show()
```

```python
write('output.wav', Fs, data)
```

```python
# Visualizing Waveform of Audio Signal
[Fs, sw] = read('sw20.wav')
# sw = sw[0:44100]
plt.plot(sw)
plt.title('Wave Plotted Samples')
plt.xlabel('Sample Number')
plt.show()
Ts = 1/Fs
t = np.arange(0, 1, Ts) # (starting_point, end_point, increment)
plt.figure()
plt.plot(t, sw)
plt.title('Wave Plotted Time')
plt.xlabel('Time (sec)')
plt.show()
```

```python
[x, Fs] = librosa.load('sw20.wav', 48000)
print('Sample Rate is : ', Fs)
plt.plot(x)
plt.title('Wave Plotted Samples')
plt.xlabel('Sample Number')
plt.show()
Ts = 1/Fs
t = np.arange(0, 1, Ts) # (starting_point, end_point, increment)
plt.figure()
plt.plot(t, x)
plt.title('Wave Plotted Time')
plt.xlabel('Time (sec)')
plt.show()
```

```python
[x, Fs] = soundfile.read('sw20.wav')
print('Sample Rate is : ', Fs)
plt.plot(x)
plt.title('Wave Plotted Samples')
plt.xlabel('Sample Number')
plt.show()
Ts = 1/Fs
t = np.arange(0, 1, Ts) # (starting_point, end_point, increment)
plt.figure()
plt.plot(t, x)
plt.title('Wave Plotted Time')
plt.xlabel('Time (sec)')
plt.show()
```
