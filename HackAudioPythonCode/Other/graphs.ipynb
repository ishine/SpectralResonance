```python
from matplotlib import pyplot as plt
import math as m
```

```python
# Plotting
x = [0, 1, 2, 3, 4]
y = [0, 1, 2, 3, 4]
plt.plot(x, y)
plt.show()
```

```python
# Stem Plot
x1 = [0, 1, 2, 3, 4]
y1 = [m.sqrt(0), m.sqrt(1), m.sqrt(2), m.sqrt(3), m.sqrt(4)]

plt.stem(x1, y1)
plt.show()
```

```python
# Using semilogx()
x2 = [1, 10, 100, 1000, 10000]
y2 = [m.log10(1), m.log10(10), m.log10(100), m.log10(1000), m.log10(10000)]

plt.semilogx(x2, y2, '-x')
plt.title('Base 10 Logarithmic Function')
plt.xlabel('x')
plt.ylabel('y(x)')
plt.show()
```

```python
# Multiple plots in one window
x3 = [0, 1, 2, 3, 4]
y3 = [m.sqrt(0), m.sqrt(1), m.sqrt(2), m.sqrt(3), m.sqrt(4)]
z3 = [pow(0, 2), pow(1, 2), pow(2, 2), pow(3, 2), pow(4, 2)]
plt.plot(x3, y3)
plt.figure()
plt.plot(x3, z3)
plt.figure(3)
plt.plot(x3, y3, x3, z3)
plt.legend(['sqrt', 'pow2'])
plt.show()
```

```python
# Subplots
x4 = [-2, -1, 0, 1, 2]
y4 = [abs(-2), abs(-1), abs(0), abs(1), abs(2)]
a = [1, 2, 3]
b = [4, 5, 6]
plt.subplot(2, 1, 1)
plt.plot(x4, y4)
plt.subplot(2, 1, 2)
plt.plot(a, b)
plt.show()
```
